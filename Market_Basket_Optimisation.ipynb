```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
          ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
          ['Dress', 'Snadals', 'Scarf', 'Pants', 'Heels'],
          ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
          ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
#Appriori

import mlxtend
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
#Support

from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support =0.6, use_colnames = True)
frequent_itemsets
```

```python
#Lift

from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.7)
```

```python
#Lift

from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.25)
```

```python
#Data loading

data = pd.read_csv("Market_Basket_Optimisation.csv")
data.fillna('0',inplace=True)
data
```

```python
#Transactions organisation

arr = []
for i in range(0,len(data)):
    arr.append([str( data.values[i,j]) for j in range(0,20) if str(data.values[i,j])!='0'])
arr = np.array(arr)
```

```python
#Apriori

te = TransactionEncoder()
data = te.fit_transform(arr)
data = pd.DataFrame(data, columns = te.columns_)
data
```

```python
#Support

frequent_itemsets_1 = apriori(data, min_support =0.05, use_colnames = True)
frequent_itemsets_1
```

```python
#Confidence

association_rules(frequent_itemsets_1, metric = 'confidence', min_threshold = 0.2)
```

```python
#Lift

association_rules(frequent_itemsets_1, metric = 'lift', min_threshold = 1.25)
```
